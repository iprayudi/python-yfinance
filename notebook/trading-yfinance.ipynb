## Pull stock data

```python
# Set label
stocks = [ # If you want to grab multiple stocks add more labels to this list
    "ICBP.JK", 
    "SIDO.JK",
#     "MYOR.JK",
#     "UNVR.JK",
#     "INDF.JK",
#     "ULTJ.JK"
] 

##
## Set start and end dates
##

# start = datetime.datetime(2020, 1, 1)
# end   = datetime.datetime(2020, 10, 10)
start = date.today()-timedelta(10)
end   = date.today()+timedelta(1) # must be D+1 to include today's data

# Grab data
df = pdr.get_data_yahoo(stocks, start = start, end = end)
print("DataFrame shape: "+ str(df.shape))
df
```
